### DATA CLEANING AND PREPROCESSING

### 5.Find missing values

```python
test.head()
```

```python
test.isnull().sum()
```

### 6.Impute missing values with mean (numerical variables)


```python
test.Loan_Amount_Term.fillna(test.Loan_Amount_Term.mean(),inplace=True)
test.LoanAmount.fillna(test.LoanAmount.mean(),inplace=True)
test.isnull().sum()
```

### 7.Impute missing values with mode (categorical variables)


```python
test.Gender.fillna(test.Gender.mode()[0],inplace=True)
test.Dependents.fillna(test.Dependents.mode()[0],inplace=True)
test.Self_Employed.fillna(test.Self_Employed.mode()[0],inplace=True)
test.Credit_History.fillna(test.Credit_History.mode()[0],inplace=True)
test.isnull().sum()
```

### 8.Remove irrelevant columns Loan_ID variable - Irrelevant from test and train data



```python
test=test.drop('Loan_ID',axis=1)
```

```python
test
```

### 9.Build dummy variables for categorical variables



```python
test=pd.get_dummies(test,drop_first=True)
test
```

### 10.Get x as columns of ApplicantIncome and Credit_History


```python
X=test.iloc[:,[0,4]]
```

```python
X
```

### 11.Import matplotlib


```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
plt.scatter(X.iloc[:, 0],X.iloc[:, 1],s=100, c='blue',label='Customer info' )
plt.title('Before Clustering')
plt.xlabel('ApplicantIncome')
plt.ylabel('Credit History')
plt.legend()
plt.show()
```

### 12.Elbow method to determine K


```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Cluster size')
plt.ylabel('WCSS')
plt.show()
```

### 13.fit K_means with founded K



```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
```

### 14.show cluster results


```python
y_kmeans
```

### 15.print cluster centers


```python
kmeans.cluster_centers_
```

### 16.Lets plot those clusters!


```python
fig = plt.figure(figsize = (15,8))
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='magenta')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='black')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],s=100,c='burlywood')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='red',label='Centroids')
plt.title('Cluster of Clients')
plt.xlabel('Applicant Income')
plt.ylabel('Credit History')
plt.legend()
plt.show()
```

### 17.Create dendogram



```python
import scipy.cluster.hierarchy as sch
fig = plt.figure(figsize = (15,10))
dendogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Applicants')
plt.ylabel('Euclidean distances')
plt.show()
```

### 18.Fit agglomerative hierarchical clustering



```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
y_hc
```

### 19.Lets plot those clusters!



```python
fig = plt.figure(figsize = (15,8))
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='magenta',label='Stars')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='blue',label='Baffled')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='green',label='Average')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='cyan',label='Target Group')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=100,c='burlywood',label='Loyals')
plt.title('Cluster of Clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.ioff()
plt.show()
```
